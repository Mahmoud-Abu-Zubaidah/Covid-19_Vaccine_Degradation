```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb
import matplotlib.pyplot as plt

def mcrmse_loss(y_true, y_pred, N=3):

    assert len(y_true) == len(y_pred)
    n = len(y_true)
    return np.sum(np.sqrt(np.sum((y_true - y_pred)**2, axis=0)/n)) / N

# Create an array of training set sizes for the learning curve
train_sizes = [int(len(X_train) * 0.1), int(len(X_train) * 0.2), int(len(X_train) * 0.5), len(X_train)]

# Initialize lists to store MCRMSE scores
train_mcrmse_scores = []
test_mcrmse_scores = []

# Train an XGBoost model and calculate MCRMSE scores for different training set sizes
for size in train_sizes:
    # Create a subset of the training data
    X_subset, y_subset = X_train[:size], y_train[:size]

    # Train the XGBoost model
    model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=600, random_state=42)
    model.fit(X_subset, y_subset)

    # Make predictions on the training set
    y_train_pred = model.predict(X_subset)
    train_mcrmse = mcrmse_loss(y_subset, y_train_pred, N=3)
    train_mcrmse_scores.append(train_mcrmse)

    # Make predictions on the validation set (X_test and y_test)
    y_test_pred = model.predict(X_test)
    test_mcrmse = mcrmse_loss(y_test, y_test_pred, N=3)
    test_mcrmse_scores.append(test_mcrmse)

# Plot the learning curve
plt.figure(figsize=(10, 6))
plt.title("Learning Curve with MCRMSE Loss")
plt.xlabel("Training Set Size")
plt.ylabel("MCRMSE Loss")
plt.grid()
plt.plot(train_sizes, train_mcrmse_scores, 'o-', label="Training MCRMSE")
plt.plot(train_sizes, test_mcrmse_scores, 'o-', label="Validation MCRMSE")
plt.legend(loc="best")
plt.show()
```
